```python
# решение 2025
class Solution(object):
    def twoSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        hashmap = {}
        for i, j in enumerate(nums):
            elem = target - j
            if elem in hashmap and i != hashmap[elem]:
                return [i, hashmap[elem]]
            hashmap[j] = i
```

```python
s = Solution()
s.twoSum([2,7,11,15], 9)
```
