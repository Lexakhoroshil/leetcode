```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

tr = TreeNode(1)
tr.left = TreeNode(7)
tr.right = TreeNode(0)
tr.left.left = TreeNode(7)
tr.left.right = TreeNode(-8)
tr.right.left = TreeNode(None)
tr.right.right = TreeNode(None)
```

```python
import copy
from typing import Optional
class Solution:
    @classmethod
    def safe_sum(cls, l):
        sum = 0
        for item in l:
            if item is not None:
                sum += item
        return sum

    def maxLevelSum(self, root: Optional[TreeNode]) -> int:
        tr_left = copy.deepcopy(root)
        tr_right = copy.deepcopy(root)
        #  
        l_left = []
        counter_left = 0
        tr_left = copy.deepcopy(root)
        while tr_left.left:
            counter_left += 1
            l_left.append(tr_left.left.val)
            l_left.append(tr_left.right.val)
            tr_left = tr_left.left

        #  

        l_right = []
        counter_right = 0
        tr_right = copy.deepcopy(root)
        while tr_right.right:
            counter_right += 1
            l_right.append(tr_right.left.val)
            l_right.append(tr_right.right.val)
            tr_right = tr_right.right

        counter_max = max(counter_right , counter_left)


        #    2d list   
        l_new = [[root.val]] + [l_left[:2]]
        for level in range(counter_max+1)[2:]:
            l_new.append(l_left[2**(level-1):2**level] + l_right[2**(level-1):2**level])

        sum_list = []

        for l in l_new:
            sum_list.append(self.safe_sum(l))

        max_sum = max(sum_list)
        return sum_list.index(max_sum) + 1
```
