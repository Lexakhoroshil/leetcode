```python
class Solution:
    def minimumAbsDifference(self, arr: List[int]) -> List[List[int]]:

        sorted_arr = sorted(arr)
        left = 0
        right = left + 1
        indexes = []
        s = abs(sorted_arr[left] - sorted_arr[right])
        s_prev = abs(sorted_arr[left] - sorted_arr[right])
        while right<len(sorted_arr):
            s_cur = abs(sorted_arr[left] - sorted_arr[right])
            if left>1:
                s_prev = abs(sorted_arr[left-1] - sorted_arr[right-1])
                if s_prev>s_cur:
                    print(left, right)
                    indexes.pop()
            s = min(s,s_cur)
            indexes.append([sorted_arr[left], sorted_arr[right]])
            #if s_cur <= s:
                

                
            right +=1
            left +=1
            
            
        return indexes
```
