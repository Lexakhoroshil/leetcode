```python
#  1161 

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

tr = TreeNode(1)
tr.left = TreeNode(7)
tr.right = TreeNode(0)
tr.left.left = TreeNode(7)
tr.left.right = TreeNode(-8)
#tr.right.left = TreeNode(None)
#tr.right.right = TreeNode(None)
```

```python
class Solution:
#       
    def maxProduct(self, root: Optional[TreeNode]) -> int:
        def recurse_total_sum(tree):
            if not tree:
                return 0
            return tree.val + recurse_total_sum(tree.left) + recurse_total_sum(tree.right)
        total_sum = recurse_total_sum(root)

        self.res = 0
        def recurse_subtree_sum(tree):
            if not tree:
                return 0
            left = recurse_subtree_sum(tree.left)
            right = recurse_subtree_sum(tree.right)
            self.res = max(self.res, (total_sum - left)*left, (total_sum - right)*right)
            return tree.val + left + right
        recurse_subtree_sum(root)
        return self.res % (10**9 + 7)
```
