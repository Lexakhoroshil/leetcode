```python
class Solution:
    def longestCommonPrefix(self, strs: List[str]) -> str:
        first_word = min(strs, key=len)
        suf = ""
        for letter in first_word:
            suf +=letter
            for word in strs:
                if not word.startswith(suf):
                    suf= suf[:-1]
                    break
            else:
                continue
            break
        return suf
```
