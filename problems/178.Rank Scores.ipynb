```python
scores = pd.DataFrame([{'id':1, 'score':3.5},
              {'id':2, 'score':3.65},
              {'id':3, 'score':4.00},
              {'id':4, 'score':3.85},
              {'id':5, 'score':4.00},
              {'id':6, 'score':3.65},
                ])
```

```python
scores[['score']].rank(method='average')
```

```python
scores[['score']].rank(method='first')
```

```python
scores[['score']].rank(method='max')
```

```python
scores[['score']].rank(method='min')
```

```python
scores[['score']].rank(method='dense')
```

```python
scores['rank'] = scores[['score']].rank(method='dense', ascending=False)
```

```python
scores.sort_values(by='rank', ascending=True).reset_index(drop=True)
```

```python
import pandas as pd

def order_scores(scores: pd.DataFrame) -> pd.DataFrame:
    scores['rank'] = scores['score'].rank(method = 'dense', ascending=False)
    return scores[['score', 'rank']].sort_values(by='rank', ascending=True)
```
