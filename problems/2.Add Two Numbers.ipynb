```python
class Solution(object):
    
    def addTwoNumbers(self, l1, l2):
        """
        :type l1: list
        :type l2: list
        :rtype: list
        """
        if l1 ==[0] and l2 == [0]:
            return [0]
        l1.reverse()
        l2.reverse()
        l3 = list(map(int, str(int(''.join([str(x) for x in l1])) + int(''.join([str(x) for x in l2])))))
        l3.reverse()
        return l3
        
```

```python
s = Solution()
s.addTwoNumbers([9,9,9,9,9,9,9], [9,9,9,9])
```

```python
s.addTwoNumbers([0], [0])
```
