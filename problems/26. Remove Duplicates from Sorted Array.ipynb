```python
class Solution:
    def removeDuplicates(self, nums: List[int]) -> int:
        left = 0
        right = left + 1
        res = [nums[0]]

        while right<len(nums):
            if nums[left] != nums[right]:
                res.append(nums[right])
            left +=1
            right +=1

        l_res = len(res)

        while len(res)<=len(nums):
            res.append('_')
        return l_res, 'nums = {}'.format(res)
```
