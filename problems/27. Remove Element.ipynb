```python
from typing import List
```

```python
nums = [3,2,2,3]
val = 3
```

```python
class Solution:
    def removeElement(self, nums: List[int], val: int) -> int:
        counter = 0
        for item in nums:
            if item != val:
                nums[counter] = item
                counter += 1
        return counter
```

```python
cv = Solution()
cv.removeElement(nums = nums, val = val)
```
