# мое решение

```python
class Solution:

    @classmethod
    def calc_uniq(cls, s):
        sub = ""
        while True:
            for i, j in enumerate(s):
                if j not in sub:
                    sub += j
                else:
                    break
            break
        return sub
    @classmethod
    def passing(cls, s):
        d = cls.calc_uniq(s)
        len_to_pass = 1 # len(d) не прошел кейс "dvdf" 0 надо еще поприкидывать - взял пока 1
        s = s[len_to_pass:]

        return s, d

    def lengthOfLongestSubstring(self, s: str) -> int:
        d_list = []
        while True:
            if s!='':
                s, d_to_list = self.passing(s)
                d_list.append(d_to_list)
            else:
                break
        if len(d_list)<1:
            return 0
        else:
            return len(max(d_list, key=len))
```
