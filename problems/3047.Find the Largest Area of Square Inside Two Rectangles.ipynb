```python
class Solution:
    @classmethod
    def calc_pair(cls, bottomLeft, topRight):
        if ((bottomLeft[0][0] >= topRight[1][0]) or (bottomLeft[0][1] >= topRight[1][1])):
            print ('first')
            return 0
        if ((bottomLeft[1][0]<topRight[0][0]) and (bottomLeft[1][1]<topRight[0][1])):
            r_bot = max(bottomLeft[0],bottomLeft[1])
            r_top = min(topRight[0], topRight[1])
            # a = r_top[1] - r_bot[0]
            # b = r_top[0] - r_bot[1]
            a = r_top[1] - r_bot[1]
            b = r_top[0] - r_bot[0]
            print('r_bot', r_bot,'r_top', r_top)
            return (min(a,b) * min(a,b) )
        else: 
            print('else')
            return 0

    def largestSquareArea(self, bottomLeft: List[List[int]], topRight: List[List[int]]) -> int:
        n = len(bottomLeft)
        res = 0
        for i in range(n):
            for j in range(i + 1, n):
                bl1, tr1 = bottomLeft[i], topRight[i]
                bl2, tr2 = bottomLeft[j], topRight[j]
                bottomLeft_r = [bl1, bl2 ]
                topRight_r = [tr1, tr2 ]
                print(bottomLeft_r)
                print(topRight_r)
                print('')
                res= max(res, self.calc_pair(bottomLeft_r,topRight_r))
                print('res = ',  res)
        return  res
```
