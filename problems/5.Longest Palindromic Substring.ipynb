```python
class Solution:
    def longestPalindrome(self, s: str) -> str:
        if len(s) <= 1:
            return s
        if 1<len(s) <= 2:
            if s != s[::-1]:
                return s[:1]

        max_s_init = 0
    
        for i, _ in enumerate(s):
            max_s = 0
            #print('----', i, '----')
            if len(s)%2 ==1:
                r = l = i
            else:
                l=i
                r=i+1
            while (l>=0 and r <len(s) and s[l] == s[r]):
                max_s = max(max_s, r-l+1)
                max_s_init = max(max_s_init,max_s)
                #print(l,r, max_s)
                if max_s == max_s_init:
                    l_fin = l
                    r_fin = r
                l -=1
                r+= 1
        # if len(s)%2 ==1:
        #     palindr = s[l_fin:r_fin+1]     
        # else:
        #     palindr = s[l_fin:r_fin+1]
        return  s[l_fin:r_fin+1]
```
