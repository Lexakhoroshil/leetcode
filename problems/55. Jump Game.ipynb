```python
class Solution:
    def canJump(self, nums: List[int]) -> bool:
        counter = 0
        if 0 not in nums[:-1]:
            return True
        elif nums[0] == 0:
            return False
        else:
            podnum = nums
            
            while (0 in podnum[:-1]):
                print(podnum)
                for i, j in enumerate(podnum):
                    if (i+j) == podnum.index(0):
                        counter +=1
                    if counter == podnum.index(0) +1:
                        return False
                    if (i+j)>podnum.index(0) and (i+j)<len(podnum):
                        podnum = podnum[:i+j]
                        counter = 0
            print(podnum)
            return True
```
