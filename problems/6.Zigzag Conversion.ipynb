```python
def convert(s: str, numRows: int) -> str:
    if numRows == 1 or numRows >= len(s):
        return s

    rows = [""] * numRows
    current_row = 0
    direction = 1

    for char in s:
        rows[current_row] += char

        if current_row == 0:
            direction = 1
        elif current_row == numRows - 1:
            direction = -1

        current_row += direction

    return rows # "".join(rows)
```

```python
def convert(s: str, numRows: int) -> str:
    if numRows == 1 or numRows >= len(s):
        return s

    cycle = 2 * numRows - 2
    result = []

    for r in range(numRows):
        for i in range(0, len(s), cycle):
            if i + r < len(s):
                result.append(s[i + r])

            if r != 0 and r != numRows - 1 and i + cycle - r < len(s):
                result.append(s[i + cycle - r])

    return "".join(result)
```
