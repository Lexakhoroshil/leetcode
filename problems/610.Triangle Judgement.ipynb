```python
import pandas as pd

def triangle_judgement(triangle: pd.DataFrame) -> pd.DataFrame:
    triangle['max_val'] = triangle[['x', 'y', 'z']].max(axis=1)
    triangle['sum'] =  triangle[['x', 'y', 'z']].sum(axis=1) -  2*triangle['max_val']
    triangle['triangle'] = 'No'
    triangle.loc[ triangle['sum'] > 0, 'triangle'] = 'Yes'
    return triangle[['x','y','z', 'triangle']]
```
