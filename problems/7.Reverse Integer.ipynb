```python
Given a signed 32-bit integer x, return x with its digits reversed. If reversing x causes the value to go outside the signed 32-bit integer range [-231, 231 - 1], then return 0.

Assume the environment does not allow you to store 64-bit integers (signed or unsigned).

 

Example 1:

Input: x = 123
Output: 321
Example 2:

Input: x = -123
Output: -321
Example 3:

Input: x = 120
Output: 21
 

Constraints:

-2**31 <= x <= 2**31 - 1
```

```python
class Solution:

    @classmethod
    def check_int32_range(cls, value):
        bits = 32
        limit = 1 << (bits - 1) 
        if -limit <= value < limit:
            return value
        else:
            return 0

    def reverse(self, x: int) -> int:

        if x==0 or not self.check_int32_range(x):
            return 0
        
        if x < 0:
            x_reversed = (str(x)[1:][::-1]).lstrip("0")
        if x > 0:
            x_reversed = (str(x)[::-1]).lstrip("0")

        s = 0
        for item in x_reversed:
            s = self.check_int32_range(s*10 + int(item))
        if x < 0:
            return s*(-1)
        else:
            return s
        
        
```
