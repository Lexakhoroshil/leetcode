```python
class Solution:
    def merge(self, nums1: List[int], m: int, nums2: List[int], n: int) -> None:
        """
        Do not return anything, modify nums1 in-place instead.
        """
        j = 0
        while j<n:
            nums1[m+j] = nums2[j]
            j+=1
        
        l = len(nums1)
        for j in range(l):
            for i in range(l-1-j):
                if nums1[i] > nums1[i+1]:
                    nums1[i], nums1[i+1] = nums1[i+1], nums1[i]
```
