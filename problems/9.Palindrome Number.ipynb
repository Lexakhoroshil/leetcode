```python
# с конвертацией в строку
class Solution:
    def isPalindrome(self, x: int) -> bool:
        if str(x) == str(x)[::-1]:
            return True
        else:
            return False
```

```python
sv = Solution()
sv.isPalindrome(122)
```

```python
# без ковертации в строку
class Solution:
    def isPalindrome(self, x: int) -> bool:
        x = str(x)
        l = len(x)
        
        if l % 2 == 0:
            part1 = x[:l//2]
            part2 = x[l//2:]
        else :
            part1 = x[:l//2]
            part2 = x[((l//2)+1):]           
        part2_rev=part2[::-1]
        if part2_rev==part1:
            return True
        else:
            return False  
```

```python
sv = Solution()
sv.isPalindrome(122)
```

```python
sv = Solution()
sv.isPalindrome(-121)
```
